```python
import netCDF4
import xarray as xr
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#1.1
# Open a netCDF4 file
ds = xr.open_dataset(r'C:\Users\zjm\Desktop\ESE_5023\200301_202006-C3S-L3_GHG-PRODUCTS-OBS4MIPS-MERGED-v4.3.nc', engine="netcdf4")#

methane_clim = ds.xch4.groupby('time.month').mean()#

methane_clim.plot(col="month", col_wrap=4, robust=True)
# 

# 
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
methane_globalmean = ds.xch4.mean(dim=['lat', 'lon'])  # 

# 
plt.figure(figsize=(12, 6))  # 
plt.plot(methane_globalmean.time, methane_globalmean, color='blue', label='Global Mean Methane')

# 
plt.xlabel("Time")  # 
plt.ylabel("Globally Averaged Methane (xCH4)")  # 
plt.title("Globally Averaged Methane Concentration Over Time")  # 

# 
plt.grid(True, linestyle='--', alpha=0.7)  # 
plt.legend()  # 

# 
plt.show()

#1881
```

```python
#1.3
methane_group = ds.xch4.groupby('time.month')#
methane_desean = methane_group - methane_group.mean() #
plt.plot(methane_desean.time,methane_desean.sel(lat= -15,lon= -150,method='nearest'))#
# 
plt.xlabel("Time")  # 
plt.ylabel("Deseasonalized methane levels")  # 
plt.title("Deseasonalized methane levels at point [15S, 150W] Over Time")  # 
#methane_desean.sel(time='2007',lat= -15,lon= -150,method='nearest')
plt.show()

#20132013
```

```python
#3.1
ds = xr.open_dataset('vocdata.nc', engine="netcdf4")   #voc1
clim_C2H3N=ds.m42_035__C2H3N___Corr_.groupby('time.hour').mean() #1
clim_C2H3N.plot()
#C2H3N9-1515
```

```python
#3.2
#plot1:C2H3N
anom_C2H3N = ds.m42_035__C2H3N___Corr_.groupby('time.hour')-clim_C2H3N
anom_C2H3N.plot()
```

```python
#plot2
ds['m42_035__C2H3N___Corr_'].plot.hist(bins=20, alpha=0.7, color='green', figsize=(8, 5))

plt.title('C2H3N Concentration Distribution')
plt.xlabel('Concentration (ppb)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#plot3
import seaborn as sns
sns.kdeplot(ds['m42_035__C2H3N___Corr_'].values, shade=True, color="orange")

plt.title('C2H3N Concentration Density Plot')
plt.xlabel('Concentration (ppb)')
plt.grid()
plt.show()
```

```python
#plot4
sns.boxplot(x=ds['m42_035__C2H3N___Corr_'].values, color='cyan')

plt.title('C2H3N Concentration Box Plot')
plt.xlabel('Concentration (ppb)')
plt.show()
```

```python
#plot5
plt.step(ds['time'], ds['m42_035__C2H3N___Corr_'], color='red', where='mid')

plt.title('C2H3N Concentration Step Plot')
plt.xlabel('Time')
plt.ylabel('Concentration (ppb)')
plt.grid()
plt.show()
```
